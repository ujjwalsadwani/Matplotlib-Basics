```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,10,20)
x
```

```python
y = np.random.randint(1,50,20)
y
```

```python
plt.plot(y)
```

```python
y = np.sort(y)
```

```python
plt.plot(y)
```

```python
plt.plot(x,y,"ko-.")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Random Plot")
plt.grid()
plt.show()
```

## Subplots

```python
plt.subplot(1,2,1)
plt.plot(x,y,"k+", markersize=5)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Random Plot")
plt.grid()
plt.subplot(1,2,2)
plt.plot(x,y,"r*", markersize=5)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Random Plot")
plt.grid()
```

```python
z = np.random.randint(1,20,20)
z = np.sort(z)
z
```

```python
fig, ax =plt.subplots()
ax.plot(x,y)
ax.plot(x,z)
#This is the basic plot you can set all the values that subplot uses i.e. label,grid etc
ax.grid()
```

## Scatter Plots

```python
plt.scatter(x,y)
```

## Bar Plots

```python
plt.bar(x,y,edgecolor="black")
```

```python
plt.barh(x,y,edgecolor="black")
```

## Histograms

```python
plt.hist(x,y,edgecolor="black")
# By default the orientation is vertical
# you can set the orientation horizontal by using the orientation property
plt.show()
```

```python
plt.hist(x,y,edgecolor="black", orientation="horizontal")
# By default the orientation is vertical
# you can set the orientation horizontal by using the orientation property
plt.show()
```

## Box Plots

```python
plt.boxplot(x,vert=True,patch_artist=True)
plt.show()
```

```python
plt.boxplot(x,vert=False,patch_artist=True)
plt.show()
```

```python
plt.stackplot(x,y)
```
