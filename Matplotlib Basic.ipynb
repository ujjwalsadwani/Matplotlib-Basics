```python
days = [1,2,3,4,5,6,7]
maxT = [50,51,52,48,47,49,46]
minT = [43,42,40,44,33,35,37]
avgT = [45,48,48,46,40,42,41]
```

```python
# plt.plot(days)
plt.plot(days,maxT, label="Max",color="black")
plt.plot(days,minT,label="Min")
plt.plot(days,avgT,label="Avg")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Weather")
plt.legend(loc="best",shadow=True,fontsize="large",facecolor="pink",edgecolor="black")
plt.grid()
plt.scatter(days,maxT)
plt.scatter(days,minT)
plt.scatter(days,avgT)
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
xpos = np.arange(len(company))
xpos
```

```python

plt.bar(xpos,revenue, label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
profit=[40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4, label="Revenue")
plt.bar(xpos+0.2,profit, width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
plt.barh(xpos-0.2,revenue, label="Revenue")
plt.barh(xpos+0.2,profit,label="Profit")

plt.yticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
```

```python
bloodSugarMen = [113,85,90,150,149,88,93,115,135,80,77,82,129]
bloodSugarWomen = [67,98,89,120,133,150,84,69,89,79,120,112,100]
```

```python
plt.hist([bloodSugarMen,bloodSugarWomen], bins=[80,100,125,150], rwidth=0.8, color = ["g","orange"], label=["Men","Women"], orientation="horizontal")
plt.xlabel("Blood Sugar Level")
plt.ylabel("Frequency")
plt.title("Blood Sugar")
plt.legend(loc="best",shadow=True, fontsize="large", facecolor="beige")
# plt.grid()
```
